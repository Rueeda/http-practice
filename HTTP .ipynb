```python
import requests


def perform_get_request():
  url = 'https://httpbin.org/get'
  response = requests.get(url)
  return response


def perform_get_request_with_params():
  url = 'https://httpbin.org/get?q=python&user=rueda'
  response = requests.get(url)
  return response



def perform_post_requests():
    url = 'https://httpbin.org/post'
    data = { 'first_name': 'Guido', 'last_name':'van Rossum'}
    response = requests.post(url , data = data)
    return response


def perform_put_requests():
    url = 'https://httpbin.org/put'
    data = {
        first_name: 'Guido',
        last_name : 'Van Rossum'

    }
    response = requests.put(url ,data = data)
    return response


def perform_patch_requests():
    url = 'https://httpbin.org/patch'
    data = {
        first_name: 'Guido',
        last_name : 'Van Rossum'
    }
    response = requests.patch(url,data=data)
    return response


def perform_delete_requests():
  url = 'https://httpbin.org/delete'
  response = requests.delete(url)
  return response


def perform_redirect_requests():
  url = 'https://httpbin.org/redirect/1'
  response = requests.get(url ,allow_redirects= True)
  return response.headers['location']

```
